```
from nltk.stem import WordNetLemmatizer
from nltk import PorterStemmer
import urllib.request
import nltk
from nltk.corpus import stopwords
nltk.download('wordnet')

# create tf function
def tf(term, token_doc):
    tf = token_doc.count(term)/len(token_doc)
    return tf

# create function to calculate how many doc contain the term 
def numDocsContaining(word, token_doclist):
    doccount = 0
    for doc_token in token_doclist:
        if doc_token.count(word) > 0:
            doccount +=1
    return doccount
  
import math
# create function to calculate  Inverse Document Frequency in doclist
def idf(word, token_doclist):
    n = len(token_doclist)
    df = numDocsContaining(word, token_doclist)
    return math.log10(n/df)
```

```
url = 'https://raw.githubusercontent.com/rutum/tf-idf/master/gutenberg/chesterton-thursday.txt'

urllib.request.urlretrieve(url, "chesterton-thursday.txt")
f = open('chesterton-thursday.txt', 'r')
text = f.read()
print(text)
```

```
url = 'https://raw.githubusercontent.com/rutum/tf-idf/master/gutenberg/chesterton-brown.txt'

urllib.request.urlretrieve(url, "chesterton-brown.txt")
f = open('chesterton-brown.txt', 'r')
text2 = text
print(text2)

```

```
lowtext = text.lower()
print(lowtext)

lowtext_2 = text2.lower()
print(lowtext_2)
```

```
nltk.download('punkt')
tokens = nltk.word_tokenize(lowtext)
print(tokens)

tokens_2 = nltk.word_tokenize(lowtext_2)
print(tokens_2)
```

```
nltk.download('stopwords')
stopwords = set(stopwords.words('english'))
filtered_token = [
    word for word in tokens if not word in stopwords and word.isalnum()]
print(filtered_token)

filtered_token_2 = [
    word for word in tokens_2 if not word in stopwords and word.isalnum()]
print(filtered_token_2)
```

```
print("###################################\r\n")
```

```
text_tf = {}
for word in filtered_token:
    text_tf[word] = tf(word, filtered_token)
# print(text_tf)
freq = nltk.FreqDist(text_tf)
print("Top 10 frequency word of Text", freq.most_common(10))
```

```
print("###################################\r\n")
```

```
text_tf_2 = {}
for word in filtered_token_2:
    text_tf_2[word] = tf(word, filtered_token_2)
# print(text_tf)
freq_2 = nltk.FreqDist(text_tf_2)
print("Top 10 frequency word of Text 2", freq_2.most_common(10))
```

```
print("###################################\r\n")
```

```
wnl = WordNetLemmatizer()
token_lem = [wnl.lemmatize(word) for word in filtered_token]
text_lem_tf = {}
for word in token_lem:
    text_lem_tf[word] = tf(word, token_lem)
freq_lem = nltk.FreqDist(text_lem_tf)
print("Top 10 frequency words in TEXT 1\r\n:", freq_lem.most_common(10))

token_lem_2 = [wnl.lemmatize(word) for word in filtered_token_2]
text_lem_tf_2 = {}
for word in token_lem_2:
    text_lem_tf_2[word] = tf(word, token_lem_2)
freq_lem_2 = nltk.FreqDist(text_lem_tf_2)
print("Top 10 frequency words in TEXT 2\r\n:", freq_lem_2.most_common(10))
```

```
doc_all = {"doc1":token_lem, "doc2":token_lem_2}
print(doc_all)
```

```
#create bag words
bag_words =[] # declare bag_words is a list
for doc in doc_all.keys():
  bag_words += doc_all[doc]
bag_words=set(bag_words)
print(bag_words)
```

```
#calculate idf for every word in bag_words
bag_words_idf={} # declare "bag_words_idf" data structure is dictionary 
for word in bag_words:
  bag_words_idf[word]= idf(word,doc_all.values())
print(bag_words_idf)
```

```
import operator

#calculate tfidf

tfidf={} # declare tfidf dictionary to store tfidf value
for doc in doc_all.keys():
  tfidf_doc={} # delare tfidf_doc as a dictionary to store tfidf of each doc
  for term in set(doc_all[doc]):
    tfidf_doc[term]= tf(term,doc_all[doc]) * bag_words_idf[term] # calculate tfidf for each doc
  # Sort tfidf by value
  tfidf_doc = dict(sorted(tfidf_doc.items(), key=operator.itemgetter(1),reverse=True))
  tfidf[doc]= tfidf_doc
```
